## Exp1: Building a simple rule based chatbot for a Laptop Retail Website.

```python
import nltk
from nltk.chat.util import Chat, reflections
```

Reflections - is a dictionary containing basic input and corresponding outputs. I added some generic term pairs such as "I","you","my" as well as certain terms specific to Laptop brand series, so that the chatbot could be made intelligent enough to suggest Laptop series to the customer without explicitly being programmed for it.  

```python
reflections = {
  "i am"       : "you are",
  "i was"      : "you were",
  "i"          : "you",
  "i'm"        : "you are",
  "i'd"        : "you would",
  "i've"       : "you have",
  "i'll"       : "you will",
  "my"         : "your",
  "you are"    : "I am",
  "you were"   : "I was",
  "you've"     : "I have",
  "you'll"     : "I will",
  "your"       : "my",
  "yours"      : "mine",
  "you"        : "me",
  "me"         : "you",
  "gamming"    : "Alienware",
  "class":"Inspiron",
  "classes":"Inspiron",
  "online classes":"Inspiron",
  "studies":"Inspiron",
  "college":"Inspiron",
  "home":"Inspiron",
  "school":"Inspiron",
  "business"   :"Vostro"
}
```

Question pairs - the first line describes the user input which we have taken as raw string input and the next line is the corresponding chatbot response. 

```python
pairs = [
    [
        r"my name is (.*)",    # Embedded variables from query in answers
        ["Hello %1, How are you today ?",]
    ],

    [   
        r"hi|hey|hello",    # Multiple query, multiple answers
        ["Hello", "Hey there",]
    ], 
    [
        r"what is your name ?",
        ["I am a bot!",]
    ],
    [
        r"(.*) services (.*) provide?",
        ["You can shop for a wide variety of laptops from us, be it gaming or simply for home.\nWe also provide Laptop care and support for all your queries.",]
    ],
    [
        r"sorry (.*)",   # One query, multiple answers
        ["Its alright","Its OK, never mind",]
    ],
    [
        r"I am fine",
        ["Great to hear that, How can I help you?",]
    ],
    [
        r"i'm (.*) doing good",   # One query, multiple answers
        ["Nice to hear that","How can I help you?:)",]
    ],
    [
        r"(.*) Is my laptop under warranty?",
        ['Please enter your service id',]
    ],
    [
        r"(.*) warranty?",
        ["You can visit our webpage to check your warranty",]
    ],
    [
        r"110(.*) warranty?",
        ["Yes, your device is under warranty",]
    ],
    [
        r"(.*) created ?",  # One query, multiple answers
        ["Shubhasmita created me using Python's NLTK library ","top secret ;)",]
    ],
    [
        r"(.*) (location|city) ?",  # multiple query, single answer
        ['We have shops in Pune, Mumbai, Delhi, Kolkata, Chennai, Bangalore',]
    ],
    [
        r"Are there shops in (Pune|Mumbai|Delhi|Kolkata|Chennai|Bangalore)?",  # multiple query, single answer
        ['Yes we have shops there!']
    ],
    [
        r"Are there any shops in (.*)?",
        ['No we don\'t have any shops there']
    ],
    [
        r"110(.*)",
        ['Yes, your device is under warranty',]
    ],
    [
        r"Name (.*) Laptops?|(.*) laptop series (.*)",   # multiple query, single answer
        ["Insipiron, XPS, Vostro, Alienware",]
    ],
    [
        r"I am looking for laptops for (.*), can you suggest?", 
        ["You can look for %1 series."]
    ],
    [
        r"Thank you",
        ["Welcome! anything else I can help you with?"]
    ],
    [
        r"No|No thank you",   # multiple query, single answer
        ["You can type 'quit' to quit"]
    ],
    [
        "quit",    # One query, multiple answers
        ["Bye take care. See you soon :) ","It was nice talking to you. See you soon :)"]
    ],
]
```

```python
def chat():
    print("Hi! I am a chatbot")
    chat = Chat(pairs, reflections)
    chat.converse()
#initiate the conversation
if __name__ == "__main__":
    chat()
```
